```python
# import deps
%matplotlib inline
from matplotlib.pyplot import plot
import numpy as np
```

```python
# load data from file
scores = []
costs = []
with open("stats.txt", "r") as f:
    for line in [_.strip() for _ in f.readlines()]:
        if line.startswith("result"):
            scores.append(int(line.split(";")[1].split(":")[1]))
        elif line.startswith("global"):
            costs.append(float(line.split(";")[1].split(":")[1]))
```

```python
plot(costs)
```

```python
plot([np.mean(x) for x in np.array_split(costs, 32)])
```

```python
plot(scores)
```

```python
plot([np.mean(x) for x in np.array_split(scores, 32)])
```

```python
plot([np.max(x) for x in np.array_split(scores, 32)])
```
